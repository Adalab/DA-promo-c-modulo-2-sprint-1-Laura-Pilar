# Estadísticas descriptivas

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datos/sephora_website_dataset.csv")
df.head()
```

### 2. Extraer los siguentes estadísiticos del conjunto de datos para varias de las variables disponibles:
- Media
- Desviación con respecto de la media
- Varianza
- Desviación típica
- Moda
- Mediana
- Media ponderada
- Robustez
- Coeficiente de variación

```python
media_rating = df["rating"].mean()
print(f"La media del rating es {media_rating}")

diferencia_rating = df["rating"] - media_rating
desviacion_media_rating = np.abs(diferencia_rating)
print(f"La desviación con respecto de la media del rating es {desviacion_media_rating}")

varianza_rating = df["rating"].var()
print(f"La varianza del rating es {varianza_rating}")

std_rating = df["rating"].std()
print(f"La desviación típica del rating es {std_rating}")

moda_rating = df["rating"].mode()
print(f"La moda del rating es {moda_rating}")

mediana_rating = df["rating"].median()
print(f"La mediana del rating es {mediana_rating}")

w = np.random.randint(0,100, df.shape[0])
avg_rating = np.average(df["rating"], weights = w)
print(f"La media ponderada del rating es {avg_rating}")

robustez = 1 / 9168
print(f"La robustez del rating es {robustez}")

coef_rating = std_rating / media_rating
print(f"El coeficiente de variación de rating es {coef_rating}")
```

```python
media_love = df["love"].mean()
print(f"La media del love es {media_love}")

diferencia_love = df["love"] - media_love
desviacion_media_love = np.abs(diferencia_love)
print(f"La desviación con respecto de la media del love es {desviacion_media_love}")

varianza_love = df["love"].var()
print(f"La varianza del love es {varianza_love}")

std_love = df["love"].std()
print(f"La desviación típica del love es {std_love}")

moda_love = df["love"].mode()
print(f"La moda del love es {moda_love}")

mediana_love = df["love"].median()
print(f"La mediana del love es {mediana_love}")

avg_love = np.average(df["love"], weights = w)
print(f"La media ponderada del love es {avg_love}")

robustez = 1 / 9168
print(f"La robustez del love es {robustez}")

coef_love = std_love / media_love
print(f"El coeficiente de variación de love es {coef_love}")
```

### 3. Interpretación de los resultados.
Ahora interpreta los resultados obtenidos de los diferentes estadísticos estudiados haciendo hincapié en aspectos interesantes de tus datos y explica lo mejor posible la naturaleza de los mismos.

- Los datos con los que hemos trabajado en la lección de hoy están extraídos de la web de venta de maquillaje online 'Sephora'.

- Las variables en las que nos hemos centrado para el análisis de hoy, son 'rating' y 'love', lo que corresponde a la puntuación que los clientes le han puesto al producto y al número de personas a quién les encanta el producto.

- Comparando la media y la mediana, vemos que en 'rating' es muy representativo, pero en 'love' nos encontramos una gran diferencia entre los dos valores. Lo que significa que en la columna 'love' habrá muchos valores atípicos.

- Relacionado con lo anterior, comparando los coeficientes de variación, vemos que en 'rating' tenemos un conjunto de datos muy homogéneo, y en cambio en 'love' nos encontramos un conjunto de datos heterogéneo.
