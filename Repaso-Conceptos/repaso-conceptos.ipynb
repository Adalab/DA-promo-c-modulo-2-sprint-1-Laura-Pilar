### Plantead una serie de preguntas como las que fuimos contestando en la clase invertida y contestad a las preguntas planteadas usando los principales estadísticos que hemos aprendido y gráficas que mejor se ajusten.

- #### Queremos saber cuántas películas se estrenaron por año.

```python
plt.figure(figsize = (10,10))

fig = sns.countplot(data = df, y = "Year", palette = "Set2")

fig.set(xlabel = "Número de películas", ylabel = "Año")

fig.set_title("Número de estrenos por año")

plt.show()
```

- En la gráfica podemos observar cómo en los años 30 y en los años 50, hubo una mayor producción de películas. Eso podría deberse a los años de bonanza que se vivieron en Estados Unidos.

- #### Queremos saber, a lo largo de los años, qué generos tuvieron más éxito.

```python
plt.figure(figsize = (10,30))

fig = sns.boxplot(data = df, y = "Genre", x = "Year")


fig.set(xlabel = "Año", ylabel = "Género")
fig.set_title("Estrenos por género y año")

plt.show()
```

```python
x = df["Genre"]
y = df["Year"]
fig, ax = plt.subplots(figsize = (20,7))

plt.scatter(x,y)

plt.xticks(rotation = 90)

plt.show()
```

- En las siguientes gráficas, podemos observar cómo los generos han ido evolucionando con los años. En la década de los 20-30 triunfaron las películas de fantasía y terror, y en los 40 y los 50, triunfó el drama por encima de cualquier otro género.
